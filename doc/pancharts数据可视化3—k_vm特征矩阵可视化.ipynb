## classwork 1

多列数据框的可视化

1. 求不同店铺的统计指标（price_mean, gmv_sum, gmv_mean, sales_mean, sales_sum, size）
2. 可视化不同店铺的统计指标

```python
a=pd.read_csv(r"e:\shoes.csv")
a["sales"]=a.sales.str.replace("人付款","").astype(int)
a["gmv"]=a.sales*a.price
```

```python
nick_tz=a.groupby("nick").agg(price_mean=("price", "mean"), 
                      gmv_sum=("gmv", "sum"),
                      gmv_mean=("gmv", "mean"),
                      sales_mean=("sales", "mean"),
                      sales_sum=("sales", "sum"),
                      size=("nick", "size"),
                      ).sort_values(by="gmv_sum", ascending=False)
nick_tz
```

```python
p4=nick_tz.head(5)
p4_vm=pc.k_vm(p4)
```

```python
tz0=p4_vm.radar()
tz0.render_notebook()
```

```python
tz0.patch_option("雷达图的里面增加阴影，透明度为0.5",verbose=True)
```

```python
tz0.render_notebook()
```

```python
tz1=p4_vm.parallel()
tz1.render_notebook()
```

```python
tz1.patch_option("并行图的每条线有不同的颜色",verbose=True)
tz1.render_notebook()
```

## rect_plot 基于dataset的直角坐标系图可视化

* 支持bar、line、scatter、effectScatter等类型的直角坐标系图可视化

* 支持一套坐标系下的多数据叠加

```python
df
```

```python
vm1.rect_plot("line",0,[1,2,3]).render_notebook()
```
