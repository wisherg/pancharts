# k2_nv 类教程

## 概述

`k2_nv` 类用于可视化 pandas 中的两层索引、单列数值型数据的序列（Series）。它可以自动将具有双层索引的 pandas Series 转换为 ECharts 可接受的数据格式，并提供关系型和三维数据可视化的图表类型。

**支持的图表类型**：
- bar3d: 3D柱状图
- graph: 图网络
- sankey: 桑基图
- heatmap: 热力图

## 1. 导入必要的模块

```python
from pancharts import k2_nv
import pandas as pd
```

## 2. 创建示例数据

我们创建一个具有两级索引的 Series，模拟产品在不同季度的销售数据：

```python
# 创建两级索引（产品 → 季度）
index = pd.MultiIndex.from_product([
    ['产品A', '产品B', '产品C', '产品D'],
    ['Q1', 'Q2', 'Q3', 'Q4']
])

# 确保数据数量与索引数量匹配（4×4=16个数据点）
values = [120, 150, 130, 180, 90, 110, 100, 130, 80, 95, 85, 110, 70, 85, 75, 95]

data = pd.Series(values, index=index, name='销售额')
data
```

```python
k2nv=k2_nv(data)
```

数据结构说明：
- 第一级：产品（产品A、产品B、产品C、产品D）
- 第二级：季度（Q1、Q2、Q3、Q4）
- 总共有 4×4 = 16 个数据点

## 3. 3D柱状图 (bar3d)

使用 `bar3d()` 方法绘制3D柱状图，展示三维数据关系：

```python
# 绘制3D柱状图
bar3d=k2nv.bar3d()
bar3d.render_notebook()
```

## patch_option与modify_option的不同

用patch_option有时只能做简单的修改，而modify_option可能可以做更复杂的修改

* patch_option是生成新的配置字典，然后通过deepmerge合并到原来的配置字典中
* modify_option是完全用大模型基于原来字典重新生成一个符合要求的配置字典

```python

bar3d.modify_option("让柱子的颜色各不相同")
bar3d.render_notebook()
```

## 4. 图网络 (graph)

使用 `graph()` 方法绘制图网络，展示节点之间的关系：

```python
# 为图网络创建数据（展示产品之间的关联）
graph_index = pd.MultiIndex.from_product([
    ['产品A', '产品B', '产品C'],
    ['用户1', '用户2', '用户3']
])

graph_values = [50, 30, 20, 40, 60, 35, 25, 45, 55]
graph_data = pd.Series(graph_values, index=graph_index, name='购买次数')
graph_data
```

```python
# 绘制图网络
chart = k2_nv(graph_data).graph()
chart.render_notebook()
```

### 使用 cate 参数指定节点分类

`graph()` 方法支持 `cate` 参数，可以为图网络的节点指定分类：

```python
# 指定节点分类：第一级为产品，第二级为用户
chart = k2_nv(graph_data, cate=['产品', '用户']).graph()
chart.patch_option("不同的cate有不同的颜色")
chart.render_notebook()
```

## 5. 桑基图 (sankey)

使用 `sankey()` 方法绘制桑基图，展示流量和流向关系：

```python
# 为桑基图创建数据（展示用户从来源到产品的流量）
sankey_index = pd.MultiIndex.from_product([
    ['搜索引擎', '社交媒体', '直接访问'],
    ['产品A', '产品B', '产品C']
])

sankey_values = [100, 80, 60, 50, 70, 40, 30, 20, 25]
sankey_data = pd.Series(sankey_values, index=sankey_index, name='访问量')
sankey_data
```

```python
# 绘制桑基图
chart = k2_nv(sankey_data).sankey()
chart.render_notebook()
```

## 6. 热力图 (heatmap)

使用 `heatmap()` 方法绘制热力图，通过颜色展示数据密度：

```python
# 使用最初的销售数据绘制热力图
chart = k2_nv(data).heatmap()
chart.render_notebook()
```

## 7. 创建另一组示例数据

让我们创建一个展示城市和月份关系的数据：

```python
# 创建两级索引（城市 → 月份）
index2 = pd.MultiIndex.from_product([
    ['北京', '上海', '广州', '深圳'],
    ['1月', '2月', '3月', '4月', '5月', '6月']
])

# 数据数量：4×6=24个
values2 = [
    25, 28, 32, 35, 30, 26,
    28, 30, 33, 36, 32, 29,
    30, 32, 35, 38, 34, 31,
    29, 31, 34, 37, 33, 30
]

data2 = pd.Series(values2, index=index2, name='平均气温')
data2
```

### 用新数据绘制3D柱状图：

```python
# 用气温数据绘制3D柱状图
chart = k2_nv(data2).bar3d()
chart.render_notebook()
```

### 用新数据绘制热力图：

```python
# 用气温数据绘制热力图
chart = k2_nv(data2).heatmap()
chart.render_notebook()
```

## 8. 使用 config 参数自定义配置

所有图表方法都支持 `config` 参数，用于传入额外的 ECharts 配置：

```python
# 自定义热力图配置
custom_config = {
    "title": {
        "text": "产品季度销售热力图",
        "left": "center"
    },
    "visualMap": {
        "min": 70,
        "max": 180,
        "calculable": True,
        "orient": "horizontal",
        "left": "center",
        "bottom": "15%"
    }
}

chart = k2_nv(data).heatmap(config=custom_config)
chart.render_notebook()
```

## 总结

`k2_nv` 类专门用于可视化具有两层索引的 pandas Series。主要特点包括：

1. 自动将双层索引转换为 ECharts 可接受的数据格式
2. 提供四种关系型和三维图表类型：
   - bar3d：3D柱状图（三维数据展示）
   - graph：图网络（节点关系展示）
   - sankey：桑基图（流量流向展示）
   - heatmap：热力图（数据密度展示）
3. 支持为图网络指定节点分类（cate 参数）
4. 支持自定义配置

**重要提示**：创建两级索引数据时，确保索引的乘积数量与值的数量一致！
