## k_vm的geo地理坐标可视化

```python
print(gk_vm.__doc__)
```

```python
#把下面这个数据转化为数据框，name为索引,后面value分为四列，分别为经度、纬度、人口、收入
z=[
        {"name": "北京", "value": [116.46, 39.92, 800,90,"市"]},
        {"name": "上海", "value": [121.48, 31.22, 900,70,"市"]},
        {"name": "广州", "value": [113.23, 23.16, 700,50,"省"]},
        {"name": "深圳", "value": [114.07, 22.62, 600,40,"省"]},
        {"name": "成都", "value": [104.06, 30.67, 500,20,"省"]}
      ]
    
z1={'经度':[],'纬度':[],'人口':[],'收入':[],'类型':[]}
z1_index=[]
for i in z:
  z1['经度'].append(i['value'][0])
  z1['纬度'].append(i['value'][1])
  z1['人口'].append(i['value'][2])
  z1['收入'].append(i['value'][3])
  z1['类型'].append(i['value'][4])
  z1_index.append(i['name'])

geo_df=pd.DataFrame(z1,index=z1_index)
geo_df
```

```python
chinamap=gk_vm(geo_df,"china")
p1=chinamap.scatter()
p1.render_notebook()
```

```python
p1=chinamap.escatter()
p1.render_notebook()
```

```python
p1=chinamap.heatmap()
p1.render_notebook()
```

```python
p1=chinamap.bar3d()
p1.render_notebook()
```

```python
p1=chinamap.scatter(3,"symbolSize")
p1.render_notebook()
```

```python
p1=chinamap.scatter([2,3],["color","symbolSize"])
p1.render_notebook()
```

```python
a=pd.read_csv(r"e:\shoes.csv")
a["sales"]=a.sales.str.replace("人付款","").astype(int)
a["gmv"]=a.sales*a.price
```

```python
city=a.location.str.split(" ",expand=True)[1]
```

```python
city_size=a.groupby(city).size()
```

```python
pd.DataFrame([{"a":"123","b":"456"}])
```

```python
import time
```

```python
city_list=[]
lng_list=[]
lat_list=[]
for i in city_size.index:
    city_list.append(i)
    z=geocode_amap(i)
    time.sleep(0.3)
    print(z)
    lng_list.append(z[0])
    lat_list.append(z[1])
```

```python
city_shoes=pd.DataFrame({'city':city_list,'lng':lng_list,'lat':lat_list,'size':city_size.values})
city_shoes.set_index('city',inplace=True)
city_shoes.head()
```

```python
city_s_p=gk_vm(city_shoes,"china")
city_s_p_s=city_s_p.scatter()
city_s_p_s.render_notebook()
```

```python
city_s_p_s=city_s_p.bar3d()
city_s_p_s.render_notebook()
```
