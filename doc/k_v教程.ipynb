# k_v 类教程

## 概述

`k_v` 类用于可视化 pandas 中的单列索引序列数据（Series）。它可以自动将 pandas Series 转换为 ECharts 可接受的数据格式，并自动判断索引类型和数值类型。

**支持的图表类型**：
- bar: 柱状图
- line: 折线图
- scatter: 散点图
- escatter: 增强散点图
- pie: 饼图
- funnel: 漏斗图
- wordcloud: 词云图
- calendar: 日历热图
- map: 地图可视化

## 1. 导入必要的模块

```python
from pancharts import k_v, Pancharts
import pandas as pd
import numpy as np
```

## 2. 创建示例数据

我们创建一个包含城市销售额的 Series 作为示例数据：

```python
# 创建数据
data = pd.Series(
    [120, 200, 150, 80, 70, 110, 130],
    index=['北京', '上海', '广州', '深圳', '杭州', '成都', '武汉'],
    name='销售额'
)
data
```

## 3. 柱状图 (bar)

使用 `bar()` 方法绘制柱状图：

```python
# 创建 kv 实例
kv = k_v(data)
```

```python
# 创建 kv 实例并绘制柱状图
chart = kv.bar()
chart.render_notebook()
```

## 4. 折线图 (line)

使用 `line()` 方法绘制折线图：

```python
# 绘制折线图
kv.line().render_notebook()
```

## 5. 散点图 (scatter)

使用 `scatter()` 方法绘制散点图：

```python
# 绘制散点图
kv.scatter().render_notebook()
```

## 6. 增强散点图 (escatter)

使用 `escatter()` 方法绘制增强散点图（带特效）：

```python
# 绘制增强散点图
chart = kv.escatter()
chart.render_notebook()
```

## 7. 饼图 (pie)

使用 `pie()` 方法绘制饼图：

```python
# 绘制饼图
chart = kv.pie()
chart.render_notebook()
```

## 8. 漏斗图 (funnel)

使用 `funnel()` 方法绘制漏斗图：

```python
# 绘制漏斗图
chart = kv.funnel()
chart.render_notebook()
```

## 9. 词云图 (wordcloud)

使用 `wordcloud()` 方法绘制词云图：

```python
# 绘制词云图
chart = kv.wordcloud()
chart.render_notebook()
```

## 10. 日历热图 (calendar)

日历热图需要时间序列索引的数据。让我们创建一个示例：

```python
# 创建时间序列数据
dates = pd.date_range(start='2024-01-01', end='2024-03-31', freq='D').strftime('%Y-%m-%d')
values = np.random.randint(0, 100, size=len(dates))
calendar_data = pd.Series(values, index=dates, name='日访问量')
calendar_data.head()
```

```python
# 绘制日历热图
chart = k_v(calendar_data).calendar()
chart.render_notebook()
```

## 11. 地图可视化 (map)

使用 `map()` 方法绘制地图，需要指定地图名称：

```python
# 创建地图数据
map_data = pd.Series(
    [1000, 800, 600, 500, 400],
    index=['北京市', '上海市', '广东省', '浙江省', '江苏省'],
    name='GDP'
)
map_data
```

```python
# 绘制中国地图
chart = k_v(map_data).map("china")
chart.render_notebook()
```

## 12. 使用 config 参数自定义配置

所有图表方法都支持 `config` 参数，用于传入额外的 ECharts 配置：

```python
# 自定义柱状图配置
custom_config = {
    "title": {
        "text": "城市销售额对比",
        "subtext": "2024年第一季度",
        "left": "center"
    },
    "tooltip": {
        "trigger": "axis",
        "axisPointer": {"type": "shadow"}
    },
    "series": [{
        "itemStyle": {
            "color": "#5470c6"
        }
    }]
}

chart = kv.bar(custom_config)
chart.render_notebook()
```

## 总结

`k_v` 类提供了丰富的图表类型来可视化单列索引的 pandas Series。主要特点包括：

1. 自动数据格式转换
2. 自动类型识别（category、value、time）
3. 支持多种常用图表类型
4. 灵活的自定义配置
5. 所有方法返回 Pancharts 实例，支持链式调用
