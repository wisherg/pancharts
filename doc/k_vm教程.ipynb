# k_vm 类教程

## 概述

`k_vm` 类用于可视化 pandas 中的单列索引、多列数值型数据的 DataFrame。它可以自动将 pandas DataFrame 转换为 ECharts 可接受的数据格式，并提供多维数据可视化的图表类型。

**支持的图表类型**：
- parallel: 平行坐标图
- radar: 雷达图
- rect_plot: 灵活数据编码方式，支持多种二维图表

## 1. 导入必要的模块

```python
from pancharts import k_vm
import pandas as pd
import numpy as np
```

## 2. 创建示例数据

我们创建一个包含产品多维度指标的 DataFrame：

```python
# 创建 DataFrame
data = pd.DataFrame(
    {
        '价格': [100, 150, 120, 200, 180],
        '销量': [500, 300, 450, 200, 250],
        '利润': [20, 35, 25, 50, 40],
        '库存': [1000, 800, 900, 600, 700],
        '评分': [4.2, 4.5, 4.0, 4.8, 4.6]
    },
    index=['产品A', '产品B', '产品C', '产品D', '产品E']
)
data
```

数据结构说明：
- 索引：产品名称（产品A-E）
- 列：5个数值维度（价格、销量、利润、库存、评分）
- 共 5 行 × 5 列 = 25 个数据点

## 实例化

```python
kvm = k_vm(data)
```

## 3. 平行坐标图 (parallel)

使用 `parallel()` 方法绘制平行坐标图，展示多维数据的趋势：

```python
# 绘制平行坐标图
chart = kvm.parallel()
chart.render_notebook()
```

## 4. 雷达图 (radar)

使用 `radar()` 方法绘制雷达图，对比各产品的综合表现：

```python
# 绘制雷达图
kvm.radar().render_notebook()
```

## 5. 灵活数据编码 (rect_plot)

`rect_plot()` 是一个非常灵活的方法，可以通过指定编码方式绘制多种二维图表。

### 基本用法：单系列散点图

```python
# 绘制散点图：价格 vs 销量
# 0 对应索引，1对应第1列（价格）
chart = kvm.rect_plot('scatter',0,1)
chart.render_notebook()
```

### 单系列折线图

```python
# 绘制折线图：产品及其价格
chart = kvm.rect_plot('line',0,2)
chart.render_notebook()
```

### 单系列柱状图

```python
# 绘制柱状图：利润与库存
chart = kvm.rect_plot('bar',3,4)
chart.render_notebook()
```

### 多系列：encode_x 为列表

注意：encode_x 为列表时，encode_y 只能为单个维度;反之 encode_y 为列表时，encode_x 只能为单个维度。

当 encode_x 为列表时，可以同时展示多个维度与同一 y 轴的关系：

```python
# 多系列折线图：价格、销量、利润 与 评分 的关系
chart = kvm.rect_plot('line',0,[4, 1, 2])
chart.render_notebook()
```

### 多系列：encode_y 为列表

当 encode_y 为列表时，可以同时展示同一 x 轴与多个维度的关系：

```python
# 多系列柱状图：价格 与 销量、利润、库存 的关系
chart = kvm.rect_plot('bar',0,[1, 2, 3])
chart.render_notebook()
```

### 混合图表类型：series_type 为列表

当 series_type 为列表时，可以为不同的系列指定不同的图表类型：

```python
# 混合图表：销量用柱状图，利润用折线图
chart = kvm.rect_plot(['bar', 'line'],[1, 2],0)
chart.render_notebook()
```

## 6. 可视化映射 (vmap_size 和 vmap_color)

`k_vm` 类提供了两个方法来配置数据的可视化映射：
- `vmap_size()`: 根据维度大小映射点的大小
- `vmap_color()`: 根据维度大小映射颜色

### 使用 vmap_size 配置点大小映射

```python
# 创建实例
vm = k_vm(data)

# 配置点大小映射：根据利润（第2列）的大小来映射点大小
size_map = vm.vmap_size(dimension=2, symbolSize=[10, 50])

# 绘制散点图
chart = vm.rect_plot('scatter', 0, 1, config=size_map)
chart.render_notebook()
```

### 使用 vmap_color 配置颜色映射

```python
# 配置颜色映射：根据库存（第3列）的大小来映射颜色
color_map = vm.vmap_color(dimension=3, color=['#50a3ba', '#eac736', '#d94e5d'])

# 绘制散点图
chart = vm.rect_plot(series_type='scatter', encode_x=0, encode_y=1, config=color_map)
chart.render_notebook()
```

### 同时使用大小和颜色映射

```python
# 合并两个映射配置
from pancharts.utils import deep_merge
combined_config = deep_merge(size_map, color_map)

# 绘制散点图
chart = vm.rect_plot(series_type='scatter', encode_x=0, encode_y=1, config=combined_config)
chart.render_notebook()
```

```python
chart = vm.rect_plot('scatter', 0, 1)
chart.dmerge(size_map).dmerge(color_map)
chart.render_notebook()
```

## 7. 创建另一组示例数据

让我们创建一个展示时间序列数据的 DataFrame：

```python
# 创建时间序列数据
dates = pd.date_range(start='2024-01-01', periods=12, freq='ME')
data2 = pd.DataFrame(
    {
        '广告支出': np.random.randint(1000, 5000, 12),
        '访客数': np.random.randint(5000, 20000, 12),
        '转化率': np.random.uniform(0.02, 0.08, 12),
        '销售额': np.random.randint(10000, 50000, 12)
    },
    index=dates.strftime('%Y-%m')
)
data2
```

### 用新数据绘制平行坐标图：

```python
# 用时间序列数据绘制平行坐标图
chart = k_vm(data2).parallel()
chart.render_notebook()
```

### 用新数据绘制雷达图：

```python
# 用时间序列数据绘制雷达图
chart = k_vm(data2).radar()
chart.render_notebook()
```

## 8. 使用 config 参数自定义配置

所有图表方法都支持 `config` 参数，用于传入额外的 ECharts 配置：

```python
# 自定义雷达图配置
custom_config = {
    "title": {
        "text": "产品综合评分雷达图",
        "left": "center"
    },
    "legend": {
        "bottom": 0
    }
}

chart = k_vm(data).radar(config=custom_config)
chart.render_notebook()
```

## 总结

`k_vm` 类专门用于可视化单列索引、多列数值的 pandas DataFrame。主要特点包括：

1. 自动将 DataFrame 转换为 ECharts 可接受的数据格式
2. 提供三种主要图表类型：
   - parallel：平行坐标图（多维趋势展示）
   - radar：雷达图（综合对比展示）
   - rect_plot：灵活数据编码（支持多种二维图表）
3. rect_plot 支持多种编码方式：
   - 单系列：encode_x 和 encode_y 都为整数
   - 多系列 x：encode_x 为列表，encode_y 为整数
   - 多系列 y：encode_x 为整数，encode_y 为列表
   - 混合类型：series_type 为列表
4. 提供可视化映射方法：
   - vmap_size：根据维度映射点大小
   - vmap_color：根据维度映射颜色
5. 支持自定义配置

**重要提示**：使用 rect_plot 时，encode_x 和 encode_y 不能同时为列表！
