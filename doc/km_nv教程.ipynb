# km_nv 类教程

## 概述

`km_nv` 类用于可视化 pandas 中的多列索引、单列数值型数据的序列（Series）。它可以自动将具有多级索引的 pandas Series 转换为树形数据结构，并提供层次化数据可视化的图表类型。

**支持的图表类型**：
- sunburst: 旭日图
- treemap: 矩形树图
- tree: 树图

## 1. 导入必要的模块

```python
from pancharts import km_nv
import pandas as pd
```

## 2. 创建示例数据

我们创建一个具有三级索引的 Series，模拟销售数据的层级结构：

```python
# 创建多级索引（3级：区域 → 城市 → 产品）
index = pd.MultiIndex.from_product([
    ['华东', '华南'],
    ['上海', '广州', '深圳'],
    ['产品A', '产品B']
])

# 确保数据数量与索引数量匹配（2×3×2=12个数据点）
values = [120, 80, 90, 70, 150, 100, 110, 95, 85, 75, 130, 115]

data = pd.Series(values, index=index, name='销售额')
data
```

数据结构说明：
- 第一级：区域（华东、华南）
- 第二级：城市（上海、广州、深圳）
- 第三级：产品（产品A、产品B）
- 总共有 2×3×2 = 12 个数据点

## 3. 旭日图 (sunburst)

使用 `sunburst()` 方法绘制旭日图，展示层级数据的占比关系：

```python
# 绘制旭日图
chart = km_nv(data).sunburst()
chart.render_notebook()
```

## 4. 矩形树图 (treemap)

使用 `treemap()` 方法绘制矩形树图，通过矩形面积展示数据大小：

```python
# 绘制完整的矩形树图
chart = km_nv(data).treemap()
chart.render_notebook()
```

### 使用 num 参数选择根节点

`treemap()` 方法支持 `num` 参数，可以只显示指定的根节点：

```python
# 只显示第一个根节点（华东）
chart = km_nv(data).treemap(num=0)
chart.render_notebook()
```

## 5. 树图 (tree)

使用 `tree()` 方法绘制树图，以树形结构展示层级关系：

```python
# 绘制树图
chart = km_nv(data).tree(0)
chart.render_notebook()
```

### 使用 num 参数选择根节点

`tree()` 方法也支持 `num` 参数：

```python
# 只显示第二个根节点（华南）
chart = km_nv(data).tree(num=1)
chart.render_notebook()
```

## 6. 创建二级索引的数据

`km_nv` 类也支持两级索引的数据。让我们创建一个示例：

```python
# 创建两级索引（部门 → 员工）
index2 = pd.MultiIndex.from_product([
    ['技术部', '市场部', '财务部'],
    ['张三', '李四', '王五']
])

# 数据数量：3×3=9个
values2 = [85, 92, 78, 88, 95, 82, 90, 87, 93]

data2 = pd.Series(values2, index=index2, name='绩效评分')
data2
```

### 使用二级索引数据绘制旭日图：

```python
# 用二级索引数据绘制旭日图
chart = km_nv(data2).sunburst()
chart.render(filename='km_nv_sunburst_level2.html')
print("二级索引旭日图已生成: km_nv_sunburst_level2.html")
```

## 7. 使用 config 参数自定义配置

所有图表方法都支持 `config` 参数，用于传入额外的 ECharts 配置：

```python
# 自定义旭日图配置
custom_config = {
    "title": {
        "text": "销售数据层级分析",
        "subtext": "按区域-城市-产品分类",
        "left": "center"
    },
    "series": [{
        "radius": [0, "90%"],
        "label": {
            "rotate": "tangential"
        }
    }]
}

chart = km_nv(data).sunburst(config=custom_config)
chart.render_notebook()
```

## 总结

`km_nv` 类专门用于可视化具有多级索引的 pandas Series。主要特点包括：

1. 自动将多级索引转换为树形数据结构
2. 自动为不同层级节点生成随机颜色
3. 提供三种层次化图表类型：
   - sunburst：旭日图（环形层级结构）
   - treemap：矩形树图（面积表示数据大小）
   - tree：树图（经典树形结构）
4. 支持通过 num 参数选择特定根节点
5. 支持自定义配置

**重要提示**：创建多级索引数据时，确保索引的乘积数量与值的数量一致！
