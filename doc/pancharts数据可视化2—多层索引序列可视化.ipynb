### classwork3

双层groupby的可视化

1. 不同'info.鞋面材质',"info.风格"的商品数量，然后按照数量倒序
2. 可视化上面结果

* 桑基图带标题
* 热力图改变颜色映射
* 网络图点变大点


```python
a=pd.read_csv(r"e:\shoes.csv")
a["sales"]=a.sales.str.replace("人付款","").astype(int)
a["gmv"]=a.sales*a.price
```

```python
p2=a.groupby(['info.鞋面材质',"info.风格"]).size().sort_values(ascending=False)
p2_k2nv=k2_nv(p2)
```

```python
heatmap0=p2_k2nv.heatmap({"title":{"text":"鞋面材质-风格-销售量","left":"center"}})
heatmap0.render_notebook()
```

```python
heatmap0.patch_option("让热力图的颜色映射变为浅绿到深绿",verbose=True)
```

```python
heatmap0.render_notebook()
```

```python
p2=a.groupby(['info.鞋面材质',"info.风格"]).size().sort_values(ascending=False)
p2_k2nv=pc.k2_nv(p2)
p2_k2nv.sankey({"title":{"text":"鞋面材质-桑基图","left":"center"}}).render_notebook()
```

```python
heatmap0=p2_k2nv.heatmap()
heatmap0.render_notebook()
```

```python
heatmap0.patch_option("让热力图的颜色映射变为浅绿到深绿",verbose=True)
heatmap0.render_notebook()
```

```python
heatmap0.option
```

```python
p2_2kvn=pc.k2_nv(p2)
graph0=p2_2kvn.graph()
graph0.render_notebook()
```

```python
graph0.patch_option("请让网络点的增大一点",verbose=True)
graph0.render_notebook()
```

```python
p2_2kvn=pc.k2_nv(p2)
graph0=p2_2kvn.bar3d()
graph0.render_notebook()
```

```python
graph0.patch_option("请让3D柱状图的柱子的颜色随高度变化",verbose=True)
graph0.render_notebook()
```

## 多层索引数值序列（km_nv）的可视化

```python
print(pc.km_nv.__doc__)
```

```python
s
```

```python
s=pd.Series([1,2,3,4,5,6],index=pd.MultiIndex.from_tuples([('a','b','c'),('a','b','d'),('a','c','e'),('a','c','f'),('b','d','g'),('b','d','h')]))
pc_s=km_nv(s)#初始化km_nv对象
pc_tree=pc_s.sunburst({
    "series": [{
        "levels": [{
            "r0": "0%",
            "r": "10%"
        }, {
            "r0": "10%",
            "r": "50%"
        }, {
            "r0": "50%",
            "r": "85%"
        }, {
            "r0": "85%",
            "r": "90%"
        }]
    }]
})#这里会基于最外层索引值个数生成若干颗树，默认选择第0颗
pc_tree.render_notebook()
```

```python
pc_tree.patch_option("通过series的levels让四层环的逐个变窄，第一层环宽度为0",verbose=True)
```

```python
pc_tree.render_notebook()
```

```python
pc_tree=pc_s.treemap()#这里会基于最外层索引值个数生成若干颗树，默认选择第0颗
pc_tree.render_notebook()
```

```python
pc_tree=pc_s.tree(0)#这里会基于最外层索引值个数生成若干颗树，默认选择第0颗
pc_tree.render_notebook()
```

```python
pc_tree.patch_option("请让树图的节点尺寸增大一点",verbose=True)
```

## classwork5

多列数据框的可视化

1. 求不同店铺的统计指标（price_mean, gmv_sum, gmv_mean, sales_mean, sales_sum, size）
2. 可视化不同店铺的统计指标

```python
nick_tz=a.groupby("nick").agg(price_mean=("price", "mean"), 
                      gmv_sum=("gmv", "sum"),
                      gmv_mean=("gmv", "mean"),
                      sales_mean=("sales", "mean"),
                      sales_sum=("sales", "sum"),
                      size=("nick", "size"),
                      ).sort_values(by="gmv_sum", ascending=False)
nick_tz
```

```python
p4=nick_tz.head(5)
p4_vm=pc.k_vm(p4)
```

```python
p4_vm.radar().render_notebook()
```

```python
p4_vm.parallel().render_notebook()
```
